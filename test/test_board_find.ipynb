```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
file='../temp/im_bgr.jpg'
img = cv2.imread(file)
# 
height, width, _ = img.shape

im2 = img[:,:,::-1]

# 2
scale_factor = 0.8

# 1 = 2.54
fig_width = width / 100 * scale_factor
fig_height = height / 100 * scale_factor

plt.figure(figsize=(fig_width, fig_height))
plt.imshow(im2)
plt.show()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)#color space conversion code.
```

```python
plt.figure(figsize=(fig_width, fig_height))
plt.imshow(hsv)
plt.show()
```

```python
title='Sabaki'
marker_b_x=20
marker_b_y=450000
marker_g_x=150
marker_g_y=100000
marker_r_x=230
marker_r_y=100000

chans = cv2.split(hsv)
colors = ("b","g","r")
plt.figure()
plt.title(title)
plt.xlabel("Bins")
plt.ylabel("# of Pixels")

for (chan,color) in zip(chans,colors):
    hist = cv2.calcHist([chan],[0],None,[256],[0,256])
    plt.plot(hist,color = color)
plt.xlim([0,256])
plt.plot(marker_b_x, marker_b_y, marker = 'v',color = 'b')
plt.plot(marker_g_x, marker_g_y, marker = 'v',color = 'g')
plt.plot(marker_r_x, marker_r_y, marker = 'v',color = 'r')
plt.show()
```

```python
lower = np.array([10, 0, 0])
upper = np.array([40, 255, 255])
mask = cv2.inRange(hsv, lower, upper)
```

```python
plt.figure(figsize=(fig_width, fig_height))
plt.imshow(mask)
plt.show()
```
